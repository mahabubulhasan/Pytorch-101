```python
## Gradient Decent

weight = 0.1
lr = 0.01

def neural_network(input, weight):
    prediction = input * weight
    return prediction

number_of_toes = [8.5]
win_or_lose_binary = [1]

input = number_of_toes[0]
true = win_or_lose_binary[0]

pred = neural_network(input, weight)

error = (pred - true)**2
print(error)

# lr = 0.1
p_up = neural_network(input, weight+lr)

e_up = (p_up - true)**2
print(e_up)


p_dn = neural_network(input, weight-lr)

e_dn = (p_dn - true)**2
print(e_dn)
```

```python
## Hot & Cold learning

weight = 0.5
input = 0.5
goal_prediction = 0.8
step_amount = 0.001

for iteration in range(1101):
    prediction = input * weight
    error = (prediction - goal_prediction) ** 2

    print(f"Error: {round(error, 2)}, Prediction: {round(prediction, 2)}")

    up_prediction = input * (weight + step_amount)
    up_error = (up_prediction - goal_prediction) ** 2

    down_prediction = input * (weight - step_amount)
    down_error = (down_prediction - goal_prediction) ** 2

    if(down_error < up_error):
        weight = weight - step_amount

    if(down_error > up_error):
        weight = weight + step_amount
```

```python
## Stopping

weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    direction_and_amount = (pred - goal_pred) * input
    weight = weight - direction_and_amount

    print(f"Error: {round(error, 3)}, Prediction: {round(pred, 3)}, Weight: {round(weight, 3)}, Direction: {direction_and_amount}")
```

```python
## Alpha: prevents overcorrecting weight updates.
weight = 0.5
goal_pred = 0.8
input = 2
alpha = 0.1

for interation in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    derivative = input * (pred - goal_pred)
    weight = weight - (alpha * derivative)

    print(f"Error: {round(error, 3)}, Prediction: {round(pred, 3)}, Weight: {round(weight, 3)}, Direction: {derivative}")
```
